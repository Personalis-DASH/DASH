##### Helper methods

```python
def get_pipeline_file_other(x):
    neo_home = x[0]
    sample = x[1]
    try:
        reporting_dir = neo_home.split('/Pipeline_Runs')[0] + '/Reports/'
        return reporting_dir + [x for x in os.listdir(reporting_dir) if (sample.split('_')[0] in x) and \
                                            (sample.split('_')[1].lower() in x.lower())][0] + '/DNA_Pipeline/Annotated_SmallVariant_Reports/All_Transcripts/{0}_somatic_dna_small_variant_report_all.xlsx'.format(sample)
    except:
        return '-'

def get_stripped_sample(x):
    if 'DNA' in x[:4]:
        return x[4:]
    else:
        return x[:-4]
    
def get_hla_file(x):
    QC_dir = x + '/QC'
    return QC_dir + '/' + [x for x in os.listdir(QC_dir) if 'HLA_type.txt' in x][0]

def get_refseq(x):
    path1 = '{0}/Transcript_Estimation/RNA_{1}/Refseq/quant.sf'.format(x[0], x[1])
    path2 = '{0}/Transcript_Estimation/{1}_RNA/Refseq/quant.sf'.format(x[0], x[1])
    path3 = '{0}/Transcript_Estimation/{1}/Refseq/quant.sf'.format(x[0], x[1])
    path4 = '{0}/Transcript_Estimation/{1}RNA/Refseq/quant.sf'.format(x[0], x[1][:-4])
    if os.path.exists(path1):
        return path1
    elif os.path.exists(path2):
        return path2
    elif os.path.exists(path3):
        return path3
    else:
        return path4
    
def get_ensembl(x):
    path1 = '{0}/Transcript_Estimation/RNA_{1}/Ensembl/quant.sf'.format(x[0], x[1])
    path2 = '{0}/Transcript_Estimation/{1}_RNA/Ensembl/quant.sf'.format(x[0], x[1])
    path3 = '{0}/Transcript_Estimation/{1}/Ensembl/quant.sf'.format(x[0], x[1])
    path4 = '{0}/Transcript_Estimation/{1}RNA/Ensembl/quant.sf'.format(x[0], x[1][:-4])
    if os.path.exists(path1):
        return path1
    elif os.path.exists(path2):
        return path2
    elif os.path.exists(path3):
        return path3
    else:
        return path4
    
def get_fusion_catcher(x):
    path = '{0}/FusionCatcher/RNA_{1}/final-list_candidate-fusion-genes.GRCh37.txt'.format(x[0], x[1])
    if os.path.exists(path):
        return path
    else:
        return 'Fail'
```

```python
def create_cluster_script(samples, normal_homes, rna_homes, neo_homes, dna_pipelines, samples_stripped, 
                          hla_types, ensembl, refseq, fusion_catcher):

    new_script_file = '/rnd/users/rpyke/code/03-public_MS/scripts/prepare_sherpa.sh'
    
    with open(new_script_file, 'w') as out_file:
        out_file.write("#! /bin/csh\n")
        out_file.write("#$ -S /bin/csh\n")
        out_file.write("#$ -cwd\n")
        out_file.write("#$ -V\n")
        out_file.write("#$ -q all.q\n")
        out_file.write("#$ -pe shm 1\n")
        out_file.write("#$ -l h_vmem=30G\n")
        out_file.write("#$ -N SHERPA\n")
        #out_file.write("#$ -t 1-2\n".format(len(samples)))
        out_file.write("#$ -t 1-{0}\n".format(len(samples)))
        out_file.write("#$ -tc 2\n")
        out_file.write("#$ -o /rnd/users/rpyke/data/03-public_MS/sge_system_files\n")
        out_file.write("#$ -e /rnd/users/rpyke/data/03-public_MS/sge_system_files\n")
        out_file.write("\n")

        out_file.write("set samples=({0})\n".format(" ".join(samples)))
        out_file.write("set normal_homes=({0})\n".format(" ".join(normal_homes)))
        out_file.write("set rna_homes=({0})\n".format(" ".join(rna_homes)))
        out_file.write("set neo_homes=({0})\n".format(" ".join(neo_homes)))
        out_file.write("set dna_pipelines=({0})\n".format(" ".join(dna_pipelines)))
        out_file.write("set samples_stripped=({0})\n".format(" ".join(samples_stripped)))
        out_file.write("set hla_types=({0})\n".format(" ".join(hla_types)))
        out_file.write("set ensembls=({0})\n".format(" ".join(ensembl)))
        out_file.write("set refseqs=({0})\n".format(" ".join(refseq)))
        out_file.write("set fusions=({0})\n".format(" ".join(fusion_catcher)))
        out_file.write("\n")

        out_file.write("set sample=$samples[$SGE_TASK_ID]\n")
        out_file.write("set normal_home=$normal_homes[$SGE_TASK_ID]\n")
        out_file.write("set rna_home=$rna_homes[$SGE_TASK_ID]\n")
        out_file.write("set neo_home=$neo_homes[$SGE_TASK_ID]\n")
        out_file.write("set dna_pipeline=$dna_pipelines[$SGE_TASK_ID]\n")
        out_file.write("set sample_stripped=$samples_stripped[$SGE_TASK_ID]\n")
        out_file.write("set hla_type=$hla_types[$SGE_TASK_ID]\n")
        out_file.write("set ensembl=$ensembls[$SGE_TASK_ID]\n")
        out_file.write("set refseq=$refseqs[$SGE_TASK_ID]\n")
        out_file.write("set fusion=$fusions[$SGE_TASK_ID]\n")
        out_file.write("\n")

        out_file.write("date\n")
        out_file.write("hostname\n")
        out_file.write("\n")
        
        #out_file.write("mkdir /rnd/users/rpyke/data/03-public_MS/inova/sherpa_input/$sample\n")
        
        out_file.write("echo python /rnd/users/mp-wsrv-hmc01/sherpa/sherpa/gather_sherpa_features.py " + \
                        "--snv_fusion_details $neo_home/Antigen_ID/tumor_peptides.details " + \
                        "--snvindel_tumor_protein_fasta $neo_home/Antigen_ID/SNVIndel/tumor_peptides_protein.fa " + \
                        #"--tumor_super_peptide_fasta $neo_home/Antigen_ID/tumor_peptides.fa " + \
                        "--ensembl_transcript_exp $ensembl " + \
                        "--refseq_transcript_exp $refseq " +\
                        "--bpseq_ffn /rnd/users/mp-wsrv-hmc01/sherpa/data/bpseq.tab " + \
                        "--hla_alleles $hla_type " + \
                        "--snvindel_all_transcript_ffn $dna_pipeline " + \
                        "--gene_prop_ffn /reference/env/dev/content/hs37d5/resources/version_control/sherpa/v1.1.0/gene_prop.tab " + \
                        "--hotspot_ffn /reference/env/dev/content/hs37d5/resources/version_control/sherpa/v1.1.0/hotspot.tab " + \
                        "--all_kmer_candidates $neo_home/Known_Antigens/all_kmer_candidates.fa " + \
                        "--fusion_catcher_ffn $fusion " + \
                        "--fasta_out /rnd/users/rpyke/data/03-public_MS/applications/NeXTDB/temp_fastas/$sample.fa " + \
                        "--sherpa_prediction_input /rnd/users/rpyke/data/03-public_MS/applications/NeXTDB/sherpa_input/$sample.tab\n")
        out_file.write("python /rnd/users/mp-wsrv-hmc01/sherpa/sherpa/gather_sherpa_features.py " + \
                        "--snv_fusion_details $neo_home/Antigen_ID/tumor_peptides.details " + \
                        "--snvindel_tumor_protein_fasta $neo_home/Antigen_ID/SNVIndel/tumor_peptides_protein.fa " + \
                        #"--tumor_super_peptide_fasta $neo_home/Antigen_ID/tumor_peptides.fa " + \
                        "--ensembl_transcript_exp $ensembl " + \
                        "--refseq_transcript_exp $refseq " +\
                        "--bpseq_ffn /rnd/users/mp-wsrv-hmc01/sherpa/data/bpseq.tab " + \
                        "--hla_alleles $hla_type " + \
                        "--snvindel_all_transcript_ffn $dna_pipeline " + \
                        "--gene_prop_ffn /reference/env/dev/content/hs37d5/resources/version_control/sherpa/v1.1.0/gene_prop.tab " + \
                        "--hotspot_ffn /reference/env/dev/content/hs37d5/resources/version_control/sherpa/v1.1.0/hotspot.tab " + \
                        "--all_kmer_candidates $neo_home/Known_Antigens/all_kmer_candidates.fa " + \
                        "--fusion_catcher_ffn $fusion " + \
                        "--fasta_out /rnd/users/rpyke/data/03-public_MS/applications/NeXTDB/temp_fastas/$sample.fa " + \
                        "--sherpa_prediction_input /rnd/users/rpyke/data/03-public_MS/applications/NeXTDB/sherpa_input/$sample.tab\n")
        out_file.write("echo Sherpa data prepared.\n")
        
                              
        out_file.write("date\n")
```

### Concating all epitopes into a single file

Getting all mutation types into a single file for each patient

```python
for model in ['BINDING', 'PRESENTATION']:
    print(model)
    for i, sample in enumerate(list(input_df.sample_name_tumor)):
        print(i, sample)
        files = [x for x in os.listdir('/rnd/users/rpyke/data/03-public_MS/applications/NeXTDB/sherpa_output') if (sample in x) and ('vMAN' in x)]
        if len(files) > 0:
            all_tmps = []
            for f in [k for k in files if model in k]:
                tmp_df = pd.read_csv('/rnd/users/rpyke/data/03-public_MS/applications/NeXTDB/sherpa_output/{0}'.format(f), sep='\t')
                all_tmps.append(tmp_df)
            all_df = pd.concat(all_tmps, sort=True)

            all_df.to_csv('/rnd/users/rpyke/data/03-public_MS/applications/NeXTDB/merged_output/{0}.{1}.all.vMAN.tab'.format(sample, model), sep='\t', index=False)
            
```
