### Importing correct purities for training/test set

```python
train_df = pd.read_csv('/rnd/users/rpyke/data/00-DASH/data/pipeline_results.training_data.v7.csv', index_col=0)
test_df = pd.read_csv('/rnd/users/rpyke/data/00-DASH/data/pipeline_results.testing_data.v7.csv', index_col=0)
```

```python
purity_ploidy_df = pd.read_csv('/rnd/users/rpyke/data/00-DASH/data/purity_ploidy.nextDB_03.csv', index_col=0)
```

```python
purity_ploidy_df['sample'] = purity_ploidy_df['Sample']
purity_ploidy_df['purity'] = purity_ploidy_df['Purity']
purity_ploidy_df['ploidy'] = purity_ploidy_df['Ploidy']
```

```python
train_df.head()
```

```python
purity_ploidy_df.head()
```

```python
len(train_df), len(test_df)
```

```python
train_df_new = pd.merge(train_df[[x for x in train_df.columns if x not in ['purity', 'ploidy']]],
         purity_ploidy_df[['sample', 'purity', 'ploidy']], on='sample', how='left')
test_df_new = pd.merge(test_df[[x for x in test_df.columns if x not in ['purity', 'ploidy']]],
         purity_ploidy_df[['sample', 'purity', 'ploidy']], on='sample', how='left')
```

```python
train_df_new.to_csv('/rnd/users/rpyke/data/00-DASH/data/pipeline_results.training_data.v8.csv')
test_df_new.to_csv('/rnd/users/rpyke/data/00-DASH/data/pipeline_results.testing_data.v8.csv')
```
